### K-means analysis
**Standardise all variable**

```python
logcol = ['Infant mortality rate','CO2 emissions','Forest area','GDP per capita']
scacol = ['Infant mortality rate','CO2 emissions','Forest area','GDP per capita','Voice and Accountability']

scaler = StandardScaler()

develop_var1 = develop_clean.copy()
develop_var1[logcol] = np.log1p(develop_var1[logcol])
scaled_data = scaler.fit_transform(develop_var1[scacol])
scaled_df = pd.DataFrame(scaled_data,columns=scacol)

scaled_df
```

**Elbow method**

```python
inertias = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=147, n_init=10)
    kmeans.fit(scaled_data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

**silhouette score and graph**

```python
fig, ax = plt.subplots(3, 2, figsize=(15,12))
for i in [2, 3, 4, 5, 6, 7]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=147)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''   
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(scaled_df)

for k in range(2,8):
    kmeans = KMeans(n_clusters=k, random_state=147, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_df)
    silhouette_avg = silhouette_score(scaled_df, cluster_labels)
    print(k, silhouette_avg)
```

```python
kmeans = KMeans(n_clusters=3, random_state=147, n_init=10)
develop_clean['Cluster3'] = kmeans.fit_predict(scaled_df)
develop_clean['Cluster3'].value_counts().sort_index()
```

```python
kmeans = KMeans(n_clusters=4, random_state=147, n_init=10)
develop_clean['Cluster4'] = kmeans.fit_predict(scaled_df)
develop_clean['Cluster4'].value_counts().sort_index()
```

```python
kmeans = KMeans(n_clusters=5, random_state=147, n_init=10)
develop_clean['Cluster_g5'] = kmeans.fit_predict(scaled_df)
develop_clean['Cluster_g5'].value_counts().sort_index()
```

```python
group_map = {0:0,1:2,2:3,3:1,4:4}
develop_clean['Cluster5'] = develop_clean['Cluster_g5'].replace(group_map)
```
